```python
import pandas as pd
f = 'transaction_data.csv'
raw = pd.read_csv(f)
```

```python
import datetime
data = raw.sample(10000)
str_type = ['UserId', 'TransactionId', 'ItemCode']
def date_time_converter(cell):
    try:
        return datetime.datetime.strptime(cell, "%a %b %d %H:%M:%S IST %Y")
    except:
        print(cell)
data.TransactionTime = data.TransactionTime.apply(date_time_converter)
data.loc[:, str_type] = data.loc[:, str_type].astype('string')
```

```python

date_condition = data.TransactionTime.map(lambda x: x.year==2018)
conditions = data.Country.eq('United Kingdom') & ~data.UserId.eq('-1') & date_condition
data[conditions]
```
